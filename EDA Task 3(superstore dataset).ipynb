```python
#importing libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(15,9))
data.groupby('Category')['Sales','Profit'].agg(['sum']).plot.bar()
plt.show()
```

```python
cor=data.corr() #correlation 
cor
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cor,annot=True,cbar=True)
```

```python
#DISTRIBUTION BASED ON REGION & CATEGOTY 
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
plt.xticks(fontsize=16,rotation=90)
sns.countplot('Region',data=data)
plt.xlabel("Region",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.title("distribution of region")
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(20,6))
plt.subplot(1,2,2)
plt.title("distribution on region based on category")
sns.countplot("Region",hue="Category",data=data)
plt.xlabel('Region',fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
#DISTRIBUTION BASED ON CATEGOTY & SUB-CATEGORY
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.xticks(fontsize=16,rotation=90)
sns.countplot('Sub-Category',data=data)
plt.xlabel("Region",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.title("distribution of category")
plt.grid(alpha=0.3)



plt.figure(figsize=(20,6))
plt.subplot(1,2,2)
plt.title("distribution of sub-category")
sns.countplot("Category",data=data)
plt.xlabel('Region',fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
#DISRTRIBUTION OF CATEGORY BASED ON SUB CATEGORY 
plt.figure(figsize=(18,10))
sns.countplot('Category',hue="Sub-Category",data=data)
plt.title("distributioin of sub-category")
plt.xlabel('sub-category')
plt.ylabel('Count')
plt.grid(alpha=0.3)
plt.show()
```

```python

#DISTRIBUTION of SEGMENTS BASED ON SHIPMENT MODE
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
sns.countplot('Segment',data=data)
plt.title('DISTRIBUTION OF SEGMENTS')
plt.xlabel('SEGMENT',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)

plt.subplot(1,2,2)
sns.countplot('Ship Mode',data=data)
plt.title('DISTRIBUTION OF SHIP MODE')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('SHIP MODE',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
    
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Discount',y='Profit',data=data)
plt.legend("Profit")
plt.grid(alpha=0.3)
plt.show()
```

```python
#DISTRIBUTION BASED ON STATES
plt.figure(figsize=(15,9))
sns.countplot('State',data=data)
plt.xticks(fontsize=16,rotation=90)
plt.xlabel("STATE",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.title("DISTRIBUTION OF STATES")
plt.grid(alpha=0.3)
plt.show()
```
